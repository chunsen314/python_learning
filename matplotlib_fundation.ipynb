```python
import matplotlib.pyplot as plt
import numpy as np
```

## **基础作图**

|函数|功能|
|---|---|
|`plt.figure(num=1, figsize=None, dpi=None)` |设置图像 |   
|`plt.x(y)label(str)`                        |设置x、y轴标签|    
|`plt.title(str)`                            |设置标题| 
|`plt.x(y)lim(top=num, bottom=num）`           |设置坐标轴区间| 
|`plt.x(y)ticks(array, name_list)`                      |设置坐标轴刻度和刻度名称|    
|`plt.legend(list or None)`                  |设置图例|


## 散点图

|参数|说明|类型|
|---|---|---|
|x，y|x, y坐标|array|
|s|点的大小|array或标量|    
|c|颜色|'r','g','b'|
|marker|标记|见文档|
|alpha|透明程度|标量|
|linewidths|点外围线宽|array或标量|
|edgecolors|点外围线颜色|标量|

```python
#生成数据集
x = np.array([0, 1, 2, 3])
y = np.array([2, 5, 7, 11])
x2 = np.arange(5)
y2 = np.array([1, 4, 3, 4, 1])
#绘图
plt.figure()                            #生成图层，可在此图层下设置不同的子图
plt.xlabel('x')                         #设置X轴标签
plt.ylabel('y')                         #设置y轴标签
plt.title('test')                       #设置标题
plt.xticks(np.arange(0, 5, step=1))     #设置X轴坐标轴与范围
plt.scatter(x, y, c='b', alpha=0.2)     #绘制散点图1
plt.scatter(x2, y2, c='b', s=80)        #绘制散点图2
plt.legend(['set1', 'set2'])            #设置图例
plt.show()
```

## **柱状图**


|参数|说明	|类型|
|---|---|---|
|x|	该系列柱状条x轴起始位置	|array|
|height	|条形的高度	|array|
|width	|宽度	|0~1，默认0.8|
|botton	|柱状条y轴起始位置	|也是y轴的起始坐标|
|align	|条形的锚定位置	|“center”中心,"edge"左边缘|
|color	|条形的颜色	|“r","b","g"|
|edgecolor|	边框的颜色|	同上|
|linewidth	|边框的宽度	|像素，默认无，int|
|tick_label	|下标的标签	|可以是元组类型的字符组合|
|log	|y轴使用科学计算法表示|	bool|
|orientation	|是竖直条还是水平条	|竖直："vertical"，水平条："horizontal"|



```python
#生成数据集
x = np.arange(1, 6)
y = np.array([0.99, 0.96, 0.97, 0.987, 0.93])
y2 = np.array([0.95, 0.93, 0.96, 0.97, 0.94])
#绘图
plt.figure()
plt.xlabel('x')
plt.ylabel('y')
plt.title('test2')
plt.bar(x-0.3, y, 0.3, align='edge', color='r', label='series1', alpha=0.5)
plt.bar(x, y2, 0.3, align='edge', color='b', label='series2', alpha=0.5)
plt.ylim(top=1, bottom=0.9)
plt.xticks(np.arange(1, 6), ['Firest', 'Second', 'Third', 'Forth', 'Fifth'])
plt.legend()
plt.show()
```

## **折线图**

|参数|说明|类型|
|---|---|---|
|x，y|x、y坐标|	array|
|color	|折线的颜色|'r','g','b'|
|marker|折线上数据点处的类型|见文档|
|linestyle|表示折线的类型|见文档|
|linewidth	|线条粗细|标量|
|alpha|表示点的透明度|0-1｜
|label|标签|str|


```python
#生成数据集
x = np.linspace(1, 5, 10)
y1 = np.random.randint(1, 10, (10))
y2 = np.array([9.8, 2.3, 3, 4.3, 5, 5.2, 3, 6, 7, 0.2])
plt.figure()
plt.xlabel('x')
plt.ylabel('y')
plt.title('test3')
plt.plot(x, y1, color='r', marker='o', linestyle='--', linewidth=1.5, label='circle')
plt.plot(x, y2, color='b', marker='*', linestyle=':', alpha=0.3, label='star')
plt.legend()
plt.show()
```
